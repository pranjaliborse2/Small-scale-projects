```python
import xarray as xr
import pandas as pd
import numpy as np
```

```python
base_url = 'https://rda.ucar.edu/thredds/dodsC/files/g/ds633.0/e5.oper.an.sfc/'
base_url2 = '/e5.oper.an.sfc.128_015_aluvp.ll025sc.'
base_url3 = '.nc'

# period of interest
pr = pd.date_range(start='2000-01',end='2015-12', freq='MS')

from datetime import timedelta
from dateutil. relativedelta import relativedelta
file_list=[]
for dt in pr:
    # get recent year and month
    year = dt.strftime('%Y%m') 
    month = dt.strftime('%Y%m%d%H')
    month2 = (dt+ relativedelta(months=1)-timedelta(1)).strftime('%Y%m%d')
    #month2 = (dt-timedelta(months=1)+ pd.DateOffset(months=1)).strftime('%Y%m%d')

    # build complete file name
    single_file=(base_url+year+base_url2 + month +'_'+ month2 +'23'+base_url3)
    file_list.append(single_file)
```

```python
ds = xr.open_dataset(file_list[1])
```
